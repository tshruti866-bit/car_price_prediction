```python
data = data.fillna("")
```

```python
data["content"] = data["text"]
```

```python
ps = PorterStemmer()
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]',' ',text) # remove all non word characters(punctuation, special characters,etc.)
    words = text.split() # split the text into individual words
    
    words =[
        ps.stem(word) # Apply stemming to each word to reduce it to its root form
        for word in words 
        if words not in stopwords.words('english') # Remove stpwords from the list
    ]
    return' '.join(words) # Join the processed words back into a single string and return it
```

```python
data["content"] =data["content"].apply(preprocess)
```

```python
print(data["content"])
```

```python
y = data["label"]
```

```python
x = data["content"]
y = data["label"]
```

```python
vectorizer = TfidfVectorizer(max_features = 5000)
x = vectorizer.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

print('x_train type:', type(x_train))
print('x_train shape:', x_train.shape)

print('y_train type:', type(y_train))
print('y_train shape:', np.array(y_train).shape)
```

```python
model = PassiveAggressiveClassifier(max_iter=50)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
```
